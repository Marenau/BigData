## Задание 6

Используя признак charges или imb, проверить, выполняется ли
центральная предельная теорема. Использовать различные длины выборок
n. Количество выборок = 300. Вывести результат в виде гистограмм. Найти
стандартное отклонение и среднее для полученных распределений.
Сделать выводы.

```python
import numpy as np

# Функция для проверки ЦПТ
def check_clt(data, sample_sizes, num_samples=300):
    results = {}
    for n in sample_sizes:
        sample_means = []
        for _ in range(num_samples):
            sample = data.sample(n, replace=True)
            sample_means.append(sample.mean())
        results[n] = sample_means
    return results

# Признаки для проверки ЦПТ
features = ['bmi', 'charges']
sample_sizes = [10, 30, 50, 100]

# Проверка ЦПТ для каждого признака
for feature in features:
    print(f"Проверка ЦПТ для признака {feature}:")
    results = check_clt(df[feature], sample_sizes)

    fig, axs = plt.subplots(2, 2, figsize=(15, 10))
    axs = axs.flatten()

    for i, n in enumerate(sample_sizes):
        axs[i].hist(results[n], bins=30, edgecolor='black')
        axs[i].set_title(f'Распределение средних значений для n={n}')
        axs[i].set_xlabel('Среднее значение')
        axs[i].set_ylabel('Частота')

        mean = np.mean(results[n])
        std = np.std(results[n])
        print(f"Для n={n}: Среднее = {mean}, Стандартное отклонение = {std}")

    plt.tight_layout()
    plt.show()
```

## Задание 7

Построить 95% и 99% доверительный интервал для среднего значения
расходов и среднего значения индекса массы тела.

```python
from scipy.stats import norm

# Функция для вычисления доверительного интервала
def confidence_interval(data, confidence_level):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = norm.ppf((1 + confidence_level) / 2)
    margin_of_error = z * (std / np.sqrt(n))
    return mean - margin_of_error, mean + margin_of_error

# Вычисление доверительных интервалов для charges
charges_95_ci = confidence_interval(df['charges'], 0.95)
charges_99_ci = confidence_interval(df['charges'], 0.99)

# Вычисление доверительных интервалов для bmi
bmi_95_ci = confidence_interval(df['bmi'], 0.95)
bmi_99_ci = confidence_interval(df['bmi'], 0.99)

print(f"95% доверительный интервал для среднего значения расходов: {charges_95_ci}")
print(f"99% доверительный интервал для среднего значения расходов: {charges_99_ci}")
print(f"95% доверительный интервал для среднего значения индекса массы тела: {bmi_95_ci}")
print(f"99% доверительный интервал для среднего значения индекса массы тела: {bmi_99_ci}")
```

## Задание 8

Проверить распределения следующих признаков на нормальность: индекс
массы тела, расходы. Сформулировать нулевую и альтернативную
гипотезы. Для каждого признака использовать KS-тест и q-q plot. Сделать
выводы на основе полученных p-значений.

```python
from scipy.stats import kstest, norm, probplot

# Функция для проверки нормальности распределения
def check_normality(data, feature_name):
    # KS-тест
    ks_stat, p_value = kstest(data, 'norm', args=(data.mean(), data.std()))
    print(f"KS-тест для признака {feature_name}:")
    print(f"Статистика KS: {ks_stat}")
    print(f"p-значение: {p_value}")

    # Q-Q plot
    plt.figure(figsize=(8, 6))
    probplot(data, dist="norm", plot=plt)
    plt.title(f'Q-Q Plot для признака {feature_name}')
    plt.xlabel('Теоретические квантили')
    plt.ylabel('Эмпирические квантили')
    plt.grid(True)
    plt.show()

    return ks_stat, p_value

# Проверка нормальности для индекса массы тела (bmi)
bmi_ks_stat, bmi_p_value = check_normality(df['bmi'], 'bmi')

# Проверка нормальности для расходов (charges)
charges_ks_stat, charges_p_value = check_normality(df['charges'], 'charges')

print(f"Для признака 'bmi': {bmi_p_value}")
print(f"Для признака 'charges': {charges_p_value}")
```

Загрузить данные из файла “bmi.csv”. Взять оттуда две выборки. Одна
выборка – это индекс массы тела людей c региона northwest, вторая
выборка – это индекс массы тела людей с региона southwest. Сравнить
средние значения этих выборок, используя t-критерий Стьюдента.
Предварительно проверить выборки на нормальность (критерий ШопироУилка) и на гомогенность дисперсии (критерий Бартлетта).

```python
df = pd.read_csv('bmi.csv')

df
```

```python
# Создание выборки для региона "northwest"
northwest_bmi = df[df['region'] == 'northwest']['bmi']

# Создание выборки для региона "southwest"
southwest_bmi = df[df['region'] == 'southwest']['bmi']

print("Northwest BMI:")
northwest_bmi
```

```python
print("Southwest BMI:")
southwest_bmi
```

```python
from scipy import stats

# Проверка на нормальность (критерий Шапиро-Уилка)
shapiro_northwest = stats.shapiro(northwest_bmi)
shapiro_southwest = stats.shapiro(southwest_bmi)

print("Шапиро-Уилк тест для northwest BMI:")
print(shapiro_northwest)

print("\nШапиро-Уилк тест для southwest BMI:")
print(shapiro_southwest)

# Проверка на гомогенность дисперсии (критерий Бартлетта)
bartlett_test = stats.bartlett(northwest_bmi, southwest_bmi)

print("\nBartlett тест для гомогенности дисперсии")
print(bartlett_test)

# Проверка на нормальность и гомогенность дисперсии
if shapiro_northwest.pvalue > 0.05 and shapiro_southwest.pvalue > 0.05 and bartlett_test.pvalue > 0.05:
    # Выполнение t-критерия Стьюдента
    t_test = stats.ttest_ind(northwest_bmi, southwest_bmi, equal_var=True)
    print("\nt-критерий Стьюдента (равные дисперсии):")
    print(t_test)
else:
    # Выполнение t-критерия Стьюдента (неравные дисперсии)
    t_test = stats.ttest_ind(northwest_bmi, southwest_bmi, equal_var=False)
    print("\nt-критерий Стьюдента (неравные дисперсии):")
    print(t_test)
```

Кубик бросили 600 раз, получили следующие результаты:

N Количество выпадений

1 97

2 98

3 109

4 95

5 97

6 104

С помощью критерия Хи-квадрат проверить, является ли полученное
распределение равномерным. Использовать функцию
scipy.stats.chisquare().

```python
from scipy.stats import chisquare

observed = np.array([97, 98, 109, 95, 97, 104])

expected = np.full(6, 600 / 6)

# Применение критерия хи-квадрат
chi2_stat, p_value = chisquare(observed, expected)

print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

# Проверка гипотезы о равномерности распределения
alpha = 0.05
if p_value > alpha:
    print("Распределение равномерное.")
else:
    print("Распределение неравномерное.")
```

```python
from scipy.stats import chi2_contingency

data = pd.DataFrame({
    'Женат': [89, 17, 11, 43, 22, 1],
    'Гражданский брак': [80, 22, 20, 35, 6, 4],
    'Не состоит в отношениях': [35, 44, 35, 6, 8, 22]
})
data.index = ['Полный рабочий день', 'Частичная занятость', 'Временно не работает', 'На домохозяйстве', 'На пенсии', 'Учёба']

# Преобразование датафрейма в таблицу сопряженности
contingency_table = data.values

# Применение критерия хи-квадрат
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("Статистика хи-квадрат:", chi2)
print("p-значение:", p_value)
print("Степени свободы:", dof)
print("Ожидаемые частоты:\n", expected)

# Проверка гипотезы о зависимости переменных
alpha = 0.05
if p_value > alpha:
    print("Семейное положение влияет на занятость.")
else:
    print("Семейное положение не влияет на занятость.")
```
